```python
import easyocr
import pandas as pd
import cv2
import matplotlib.pyplot as plt
```

```python
reader = easyocr.Reader(['id'],False,)
```

```python
hasil = reader.readtext('Image/example02.jpg')
hasil
```

```python
df_hasil = pd.DataFrame(hasil,columns=['BBOX','Text','Confident'])
df_hasil
```

```python
img = cv2.imread('Image/example02.jpg')
```

```python
df_hasil.BBOX[0][0]
```

```python
df_hasil.BBOX[0][1]
```

```python
list(map(int,df_hasil.BBOX[0][0]))
```

```python

for i in range(df_hasil.shape[0]):
    # print(i)
    img = cv2.rectangle(img,pt1=list(map(int,df_hasil.BBOX[i][0])),pt2=list(map(int,df_hasil.BBOX[i][2])),color=(255,0,0))
plt.imshow(img)
```
